# XOR

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt 
```

```python
# 1. data set
x_data = [[0,0], [0,1], [1,0], [1,1]]
y_data = [[0],   [1],   [1],   [0]]

# 2.   
model = Sequential()
model.add(Dense(10, input_dim=2, activation='relu'))
model.add(Dense(20, activation='relu')) #  input_dim  .
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

#  2,  10 : Weight 20(10*2), bias 10 
print(model.summary())

# 3.   
model.compile(loss='mse', optimizer='adam', metrics=['binary_accuracy'])
# model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=['accuracy'])
# model.compile(loss="binary_crossentropy", optimizer="adam", metrics=['accuracy'])

# 4.  
fit_hist = model.fit(x_data, y_data, epochs=500)
```

```python
fit_hist.history.keys()
```

```python
for acc in fit_hist.history['binary_accuracy']:
    print(acc, end='\t')
```

```python
for loss in fit_hist.history['loss']:
    print(loss, end='\t')
```

```python
plt.plot(fit_hist.history['loss'])
plt.show()
```

```python
#   '0 0'
while True:
    input_list = input('space  0 1 2 (:9)').strip().split()
    input_data = list(map(int, input_list))
    if input_data[0] == 9:
        break;
    input_data = np.array(input_data).reshape(1,2)
    
    print(': ', input_data)
    print(': ', int(model.predict(input_data).round()))
```

# OR

```python
# 1. data set
x_or_data = np.array([[0,0],[0,1],[1,0],[1,1]], 'float32')
y_or_data = np.array([[0],[1],[1],[1]] , 'float32')

# 2.  
model = Sequential()
model.add(Dense(32, input_dim=2, activation='relu'))
model.add(Dense(1, activation="sigmoid"))

# 3.  
model.compile(loss="mse", optimizer="adam", metrics=['binary_accuracy'])
# model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=['accuracy'])
# model.compile(loss="binary_crossentropy", optimizer="adam", metrics=['accuracy'])

# 4. 
fit_hist = model.fit(x_or_data, y_or_data, epochs=500, verbose=1)
```

```python
plt.plot(fit_hist.history['loss'])
```

# AND

```python
# 1. data set
x_and_data = np.array([[0,0],[0,1],[1,0],[1,1]], 'float32')
y_and_data = np.array([[0],[0],[0],[1]] , 'float32')

# 2.  
model = Sequential()
model.add(Dense(32, input_dim=2, activation='relu'))
model.add(Dense(1, activation="sigmoid"))

# 3.  
model.compile(loss="mse", optimizer="adam", metrics=['binary_accuracy'])

# 4. 
fit_hist = model.fit(x_and_data, y_and_data, epochs=500, verbose=1)
```

```python
plt.plot(fit_hist.history['loss'])
```
