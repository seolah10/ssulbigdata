```python
import pandas as pd 
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler # scale 조정 
from sklearn.model_selection import train_test_split # 데이터 분리 
from tensorflow.keras import metrics
# 성능평가지표
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
```

```python
# 1. 데이터 셋 불러오기 
raw_data = pd.read_excel('data/heart-disease.xlsx')
raw_data.head(3)
```

```python
clean_data = raw_data.replace("?", np.nan)
clean_data
```

```python
clean_data = clean_data.dropna() # 결측치가 있는 행 제거 
clean_data
```

```python
clean_data.info()
```

```python
clean_data.isnull().sum()
```

```python
clean_data.columns
```

```python
input_ = clean_data.iloc[:, :-1]
target_ = clean_data.iloc[:, [-1]]
target_
```

```python
# 심장병 걸린 데이터 수 
target_['heartDisease'].sum() # 값이 0 혹은 1이므로 더하면 됨 
```

```python
# 심장병 걸린 데이터 비율 
target_['heartDisease'].mean()
```

```python
input_.head(2)
```

```python
# scale 조정 
scaler = StandardScaler()
scaler.fit(input_)
scaled_input = scaler.transform(input_)
scaled_input = pd.DataFrame(scaled_input)
scaled_input
```

```python
scaled_input.describe().T
```

```python
# 학습 데이터 : 테스트 데이터 = 7:3 (비율은 자유설정이지만, 보통은 이렇게함)
x_train , x_test, y_train, y_test = train_test_split(
    scaled_input, target_, test_size=0.3, random_state=5)
print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
```

```python
# 2. 모델 구성 
model = Sequential()
model.add(Dense(units=500, input_dim=13, activation='tanh'))
model.add(Dropout(0.1))
model.add(Dense(units=200, activation='tanh'))
model.add(Dropout(0.1))
model.add(Dense(units=100, activation='tanh'))
model.add(Dropout(0.1))
model.add(Dense(units=1, activation='sigmoid'))
model.summary()
```

```python
# 3. 학습과정 설정 
model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
# model.compile(loss="categorical_crossentropy", optimizer="nadam", 
#              metrics=[metrics.Recall()])
#model.compile(loss="mse", optimizer="adam",  
#              metrics=[metrics.Precision()])
```

- accuracy(정확도) : 예측결과와 실제값이 동일한 건수 / 전체 건수 
- Recall(재현율) : 실제 True인데 예측도 True로 한 건수 / 실제 True인 전체 건수
- Precision(정밀도) : 실제True인데 예측도 True로 한 건수 / True로 예측한 건수

```python
# 4.학습시키기 
hist = model.fit(x_train, y_train, 
                 epochs=50, batch_size=50,
                 validation_split=0.2, verbose=2)
```

```python
# 학습과정 표시 
fig, loss_ax = plt.subplots()

loss_ax.plot(hist.history['loss'], 'y', label='train loss')
loss_ax.plot(hist.history['val_loss'], 'g', label='val loss')
loss_ax.set_xlabel('epochs')
loss_ax.set_ylabel('loss')

acc_ax = loss_ax.twinx() # x축 공유 
acc_ax.plot(hist.history['accuracy'], 'b', label='train accuracy')
acc_ax.plot(hist.history['val_accuracy'], 'r', label='val accuracy')
acc_ax.set_ylabel('accuracy')

loss_ax.legend(loc='upper left')
acc_ax.legend(loc='lower left')
```

```python
# 5. 평가 및 예측 
score = model.evaluate(x_test, y_test, verbose=0)
print('model loss: ', score[0])
print('model accuracy', score[1])
```

```python
# 6. confusion_matrix(성능평가 지표)
pred = model.predict(x_test)
pred = (pred > 0.5)

print(confusion_matrix(y_test, pred), end='\n')
print('f1_score: ', f1_score(y_test, pred))
```
