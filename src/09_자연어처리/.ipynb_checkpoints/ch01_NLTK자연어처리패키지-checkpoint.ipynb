```python
emma = nltk.corpus.gutenberg.raw('austen-emma.txt')
print(emma[:20])
```

```python
len(emma)
```

```python
# 문장 단위로 쪼개기 
from nltk.tokenize import sent_tokenize
sent_tokens = sent_tokenize(emma)

print('문장 수: ', len(sent_tokens))
print('%r' % (sent_tokens[10]))
```

```python
# 단어 단위로 쪼개기 
from nltk.tokenize import word_tokenize
word_tokenize(sent_tokens[10])
```

```python
# print(sent_tokens[0]) # 곳곳에 특수문자들이 있는 것을 볼 수 있다. 

# RegexpTokenizer 클래스: 토큰화할 때 정규표현식 이용 
from nltk.tokenize import RegexpTokenizer
ret = RegexpTokenizer('[\w]+') # [\w] = [0-9a-zA-Z] / +: 한 개 이상 
print(ret.tokenize(sent_tokens[10])) # 특수문자가 사라짐 
```

# 형태소 분석 
- 형태소: 의미가 있는 가장 작은 말의 단위 
     - 어간 추출(Stemming)
     - 원형 복원(Lemmatizing)
     - 품사 태깅(Part of Speech Tagging)

```python
words = ['sending', 'cooking', 'files', 'lives', 'crying', 'dying']
```

```python
# 어간 추출 (1) : PorterStemmer
from nltk.stem import PorterStemmer

pst = PorterStemmer()
pst.stem(words[0]), pst.stem(words[1]), pst.stem(words[2]) 
```

```python
[pst.stem(word) for word in words]
```

```python
# 어간 추출 (2) : LancasterStemmer
from nltk.stem import LancasterStemmer

lst = LancasterStemmer()
[lst.stem(word) for word in words]
```

```python
# 어간 추출 (3) : RegexpStemmer
from nltk.stem import RegexpStemmer

rst = RegexpStemmer('ing')
[rst.stem(word) for word in words]
```

```python
# 어간 추출을 하면 의미가 달라질 수 있어 원형복원을 한다.
words2 = ['believe', 'cooking']
[lst.stem(word) for word in words2]
```

```python
# 원형 복원 
from nltk.stem.wordnet import WordNetLemmatizer

wl = WordNetLemmatizer()
[wl.lemmatize(word) for word in words2]
```

```python
# 품사 태깅 
from nltk.tag import pos_tag

tagged_list = pos_tag(word_tokenize(sent_tokens[10])) # 11번째 문장을 단어로 토크나이징 
print(word_tokenize(sent_tokens[10]))
print('\n품사 태깅 결과')
print(tagged_list) # but 결과를 100% 신뢰할 수는 없다.
```

**<font size='4'>QUIZ: 소설 [Emma]에서...</font>** 
1. 특수문자가 들어가지 않은 3글자 이상의 단어만 추출에서 품사 태깅
2. Emma 단어의 품사 태깅이 어떤 품사들로 되어 있는지 모두 출력 
3. 명사의 단어만 뽑아 등장하는 명사의 종류 수 출력 

```python
# 1.
ret = RegexpTokenizer('[\w]{3,}')

emma_tags = pos_tag(ret.tokenize(emma))
emma_tags
```

```python
# 2. 
pos = set() # 중복 제거를 위해 집합 사용 
cnt = 0

for emma_t in emma_tags:
    if emma_t[0] == 'Emma':
        cnt += 1 
        pos.add(emma_t[1]) # set()함수는 append가 아닌 add를 사용함 
print(cnt)
print(pos)
```

```python
# 위에서 출력된 품사별 갯수 출력
import pandas as pd # 시리즈 생성 

pos_cnt = pd.Series([0]*len(pos), index=list(pos))
for emma_t in emma_tags:
    if emma_t[0] == 'Emma':
        pos_cnt[emma_t[1]] += 1
pos_cnt
```

```python
# 3. 
# 방법 1 - 리스트 컴프리헨션 
nouns_list = [emma_t[0] for emma_t in emma_tags if emma_t[1] == 'NN']

print('명사가 나온 횟수: ', len(nouns_list))
print('나오는 명사의 갯수: ', len(set(nouns_list))) # 중복되는 명사 제거 
print('한 단어가 나오는 평균 빈도 수: ', len(nouns_list) / len(set(nouns_list)))
```

```python
# 방법 2 - for문 이용
nouns_list = []

for emma_t in emma_tags:
    if emma_t[1] == 'NN':
        nouns_list.append(emma_t[0])
len(nouns_list)
```

```python
# 3. 명사들만 추출 
nouns_list = [emma_t[0] for emma_t in emma_tags if emma_t[1] == 'NN']
nouns_list
```
