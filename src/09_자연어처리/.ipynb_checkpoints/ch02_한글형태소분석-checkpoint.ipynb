## Hannanum

```python
from konlpy.tag import Hannanum

han = Hannanum()
han.analyze(text)
```

https://konlpy.org <br>

```python
# 형태소 분석만 
print(han.morphs(text))
```

```python
# 품사 태깅
print(han.pos(text, ntags=9)) # default: ntags=9 (9가지 품사로 나뉘어서 태깅서 태깅)
```

```python
print(han.pos(text, ntags=22))
```

```python
# 형용사(PA)만 추출 - for문 
tagged_text = han.pos(text, ntags=22)
l = []

for t in tagged_text:
    if t[1] == 'PA':
        l.append(t[0])
l
```

```python
# 형용사(PA)만 추출 - 리스트 컴프리헨션
[t[0] for t in tagged_text if t[1] == 'PA']
```

```python
# 명사만 추출
han.nouns(text)
```

## Kkma

```python
from konlpy.tag import Kkma

kkma = Kkma(max_heap_size=1024) # max_heap_size: heap memory size를 증가시킬 때 사용
print(kkma.morphs(text)) # 형태소 분석 
```

```python
print(kkma.pos(text, 56)) # default: ntags=30
```

```python
# 보통명사만 추출 
tagged_text = kkma.pos(text, 56)
[t[0] for t in tagged_text if t[1] == 'NNG']
```

```python
# 모든 명사 추출 
kkma.nouns(text)
```

# 워드 임베딩 
    : 단어간 유사성 도출

```python
# 뉴스 기사 link들 list 가져오기 
import requests 
from bs4 import BeautifulSoup

rss_url = "https://fs.jtbc.joins.com/RSS/economy.xml"
jtbc_economy = requests.get(rss_url)
economy_soup = BeautifulSoup(jtbc_economy.content, 'xml') 
                                                # 'html.parser'라고 해도됨
link_list = economy_soup.select('item > link')
link_list = [link.text for link in link_list]
link_list
```

```python
len(link_list)
```

```python
# 뉴스 기사들의 명사들만 뽑아 워드 임베딩하기 편하게 만들기 
kkma = Kkma()
news = []

for link in link_list:
    news_reponse = requests.get(link)
    news_soup = BeautifulSoup(news_reponse.content, 'html.parser')
    news_title = news_soup.select_one('h3#jtbcBody')
    news_content = news_soup.select_one('div#articlebody > div.article_content')
    
    content = news_title.text + ' ' + news_content.text
    news.append(kkma.nouns(content))
```

```python
# 기사별 명사들의 갯수
print([len(n) for n in news])
```

```python
# 워드 임베딩 
from gensim.models import Word2Vec # pip install gensim
model = Word2Vec(news, 
                 size=100,
                 window=5,
                 min_count=2,
                 wordkers=-1)
```

```python
# 코로나와 가장 유사성이 있다고 여겨지는 단어들 (거리 기준)
model.wv.most_similar('코로나')
```

```python
model.wv.most_similar('코로나', topn=5)
```
